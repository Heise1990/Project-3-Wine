```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine_data = pd.read_csv('/Users/marinaduarte/Documents/Project-3-Wine/winemag-data_first150k.csv')
```

```python
wine_data
```

```python
for_points = wine_data.dropna(subset=['points'])
for_points
```

```python
#points = wine_data['points'].astype(int) # or str for string
for_points = for_points.astype({"points": int})
```

```python
for_points.dtypes
```

```python
country_points = for_points.groupby(['country'])['points'].mean()
#country_points2 = country_points.to_frame()
#country_points2['points'].unstack()
#country_points2.sort_values(by='points', ascending = False)
country_points
```

```python
country_points.plot.bar(figsize = [20,10])
```

```python
pointsDF = for_points.merge(country_points, how='inner', left_on=["country"], right_on=["country"])
pointsDF
```

```python
#points = wine_data['points'].astype(int) # or str for string
for_price = wine_data.dropna(subset=['price'])
for_price = for_price.astype({"price": int})
for_price.head()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.bar(pointsDF['country'], pointsDF['points_y'])
plt.xticks(rotation=90)
ax.set_ylim([80,95])
plt.title('Average points by country')
plt.show()
```
